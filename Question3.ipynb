```python
import pandas as pd
import numpy as np
from skmultilearn.problem_transform import ClassifierChain
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn import neighbors
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_predict
from sklearn.metrics import classification_report, confusion_matrix, roc_curve, auc, hamming_loss
import scikitplot as skplt
import matplotlib.pyplot as plt
import warnings

def logistic_regression(X, y):
    cross_fold_vals = [5, 10]
    c_values = [1.0, 2.0, 5.0, 10.0]
    for cf_val in cross_fold_vals:
        for c in c_values:
            logreg = LogisticRegression(random_state=0, solver='lbfgs', C=c)
            clf = ClassifierChain(logreg)
            y_pred = cross_val_predict(clf, X, y, cv=cf_val)
            evaluations('Logistic Regression with C = ' + str(c), y, y_pred)
    return

def random_forest(X, y):
    cross_fold_vals = [5, 10]
    estimator_values = [10, 20, 50, 100]
    for cf_val in cross_fold_vals:
        for est in estimator_values:
            rf = RandomForestClassifier(n_estimators=est)
            clf = ClassifierChain(rf)
            y_pred = cross_val_predict(clf, X, y, cv=cf_val)
            evaluations('Random Forest with estimator value ' + str(est), y, y_pred)
    return

def support_vector_classifier(X, y):
    cross_fold_vals = [5, 10]
    kernal_types = ['linear', 'rbf', 'sigmoid']
    for cf_val in cross_fold_vals:
        for kernal in kernal_types:
            svclassifier = SVC(kernel=kernal)
            clf = ClassifierChain(svclassifier)
            y_pred = cross_val_predict(clf, X, y, cv=cf_val)
            evaluations('SVM with ' + str(kernal) + ' kernal', y, y_pred)
    return

def knn_classifier(X, y):
    cross_fold_vals = [5, 10]
    neighbors_k_vals = [1, 2, 5, 10, 20]
    for cf_val in cross_fold_vals:
        for k in neighbors_k_vals:
            knn = neighbors.KNeighborsClassifier(n_neighbors = k)
            clf = ClassifierChain(knn)
            y_pred = cross_val_predict(clf, X, y, cv=cf_val)
            evaluations('KNN with k=' + str(k), y, y_pred)
    return
    
def evaluations(title, y_test, y_pred):
    print('Performance results for ', title)
    print('Hamming Loss: ', hamming_loss(y_test, y_pred))
    print(classification_report(y_test, y_pred))
    skplt.metrics.plot_roc_curve(y_test[0], get_dummies(y_test[0], y_pred.toarray()[:,0]))
    plt.show()
    print('###############################################################\n')
    return

def get_dummies(y_test, y_pred):
    xx = np.unique(y_test)
    condlist = [y_pred==xx[0], y_pred!=xx[0]]
    choicelist = [1, 0]
    t = np.select(condlist, choicelist)
    for i in range(1, len(xx)):
        condlist = [y_pred == xx[i], y_pred != xx[i]]
        t = np.vstack((t, np.select(condlist, choicelist)))
    return np.transpose(t)
        

def main():
    train_dataset = '../NewDataset/MultLabelTrainData.txt'
    test_dataset = '../NewDataset/MultLabelTestData.txt'
    labels_path = '../NewDataset/MultLabelTrainLabel.txt'
    train_df = pd.read_csv(train_dataset, sep='\t', header=None)
    test_df = pd.read_csv(test_dataset, sep='\t', header=None)
    train_labels_df = pd.read_csv(labels_path, sep='\t', header=None)
    
    logistic_regression(train_df, train_labels_df)
    random_forest(train_df, train_labels_df)
    support_vector_classifier(train_df, train_labels_df)
    knn_classifier(train_df, train_labels_df)
    
if __name__ == '__main__':
    warnings.filterwarnings('ignore')
    main()
```
